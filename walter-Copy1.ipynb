```python
# Read in data and convert index to a datetime
wBook_load = "wBooks.csv"
wBooks = pd.read_csv(wBook_load)
wBooks = pd.DataFrame(wBooks) 
wBooks.dtypes
```

```python
# wBooks.to_csv('wBooks.csv', index=False)
```

```python
wBooks["sqrt_pages"] = (wBooks['num_pages']**0.5) 


wBooks["sizerating^2"] = wBooks["average_rating"]**2 

wBooks["size0"] = wBooks["z_score"] + 8.015 
wBooks["sizeZ0^2"] = wBooks["size0"]**2
wBooks["size5"] = wBooks["z_score"]*5 
wBooks["size10"] = wBooks["z_score"]*10 

wBooks["size^2"] = np.power((wBooks["z_score"]),2) 
wBooks["size^3"] = np.power((wBooks["z_score"]),3) 

wBooks["sizesqrtpages"] = (wBooks['num_pages']**0.5) 
wBooks["sizeRevRate"] = wBooks["revRate"]*100 
```

```python
import plotly.express as px


fig = px.scatter(wBooks, 
                 x="log_ratings", 
                 y="log_reviews", 
                 hover_name="title", 
                 hover_data=["average_rating"], 
                 size=wBooks['revRate'])                 
fig.show()
```

```python
wBooks.iplot(
    x='num_pages',
    y='z_score',
    opacity=0.4,
    size=wBooks['sizeRevRate'],
    mode='markers',
    layout=dict(
        xaxis=dict(title='number of pages'),
        yaxis=dict(title='z-score of average review'),
        title='Marker size is likelihood of text review'))
```
